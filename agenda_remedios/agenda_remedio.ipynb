##### Organiza horário dos remédios

```python
from datetime import datetime, timedelta
from collections import defaultdict

def formata_data(data_):
  return datetime.strptime(data_, "%d/%m/%Y %H:%M")

def calcular_horarios(nome, atual, fim, freq, agenda):
  if atual > fim:
    return
  agenda[atual].append(nome)
  calcular_horarios(nome, atual + timedelta(hours=freq), fim, freq, agenda)

def horarios_remedios(nomes, inicios, periodos, frequencias):
  agenda = defaultdict(list)

  if len(inicios) == 1:
    inicios = inicios * len(nomes)

  inicios = [formata_data(i) for i in inicios]

  for nome, inicio_str, dias, freq in zip(nomes, inicios, periodos, frequencias):
    fim = inicio_str + timedelta(days=dias)
    calcular_horarios(nome, inicio_str, fim, freq, agenda)

  agenda_ordenada = dict(sorted(agenda.items()))
  return agenda_ordenada


nomes = ["amoxicilina", "ibuprofeno", "prednisona", "dexclorfeniramina"]
inicios = ["14/04/2025 16:30", "14/04/2025 16:30", "14/04/2025 18:30", "14/04/2025 18:30"]
periodos = [7, 5, 5, 5]
frequencias = [8, 5, 12, 8]

agenda = horarios_remedios(nomes, inicios, periodos, frequencias)

for horario, remedio in agenda.items():
  print(horario.strftime("%d/%m/%Y %H:%M"), "->", ", ".join(remedio))
```
