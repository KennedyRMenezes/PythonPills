```python
import pandas as pd
import numpy as np
```

```python
num_linhas = 100

ids = np.arange(1, num_linhas + 1)
nomes = [f"Pessoa_{i}" for i in ids]
idades = np.random.randint(18, 60, size=num_linhas)
notas = np.round(np.random.uniform(0, 10, size=num_linhas), 2)

df = pd.DataFrame({
    "id":  ids,
    "nome": nomes,
    "idade": idades,
    "notas": notas
})
```
